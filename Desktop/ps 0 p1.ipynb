```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
```

```python
# part a
train_data = np.loadtxt('https://raw.githubusercontent.com/lmu-mandy/ps0-grantgambetta/master/lm-data/brown-train.txt?token=AINEJ23FRNFROKYKRFCGQWC7MPTO2', dtype=str)
print(train_data[0:20])
```

```python
tokenized_train_data = word_tokenize(train_data)
print(tokenized_train_data)
```

```python
# part b
unique, counts = np.unique(train_data, return_counts=True)
counts = dict(zip(unique, counts))
sort_orders = sorted(counts.items(), key=lambda x: x[1], reverse=True)
for i in sort_orders:
	print(i[0], i[1])
```

```python
counts_df = pd.DataFrame(sort_orders)
counts_df.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(counts_df[0], counts_df[1])
plt.xticks([])
plt.show()
```
